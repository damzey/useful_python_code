```python
import matplotlib.pyplot as plt
```

```python
x = [3, 5, 7, 9, 11]
y = [5, 3, 7, 8, 2]

plt.scatter(x, y)
plt.show()
```

```python
x = [3, 5, 7, 9, 11]
y = [5, 3, 7, 8, 2]
labels = ["Jan", "Feb", "Mar", "April", "May"]
plt.scatter(x, y)
for i, j in enumerate(labels):
    plt.annotate(j, (x[i]+0.10, y[i]), fontsize=10)
plt.show()
```
